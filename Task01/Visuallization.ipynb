```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#LOADING RANDOM DATA SET
sns.get_dataset_names()
```

```python
tips=sns.load_dataset("tips")
iris=sns.load_dataset("iris")
planets=sns.load_dataset("planets")
titanic=sns.load_dataset("titanic")
```

```python
tips
```

```python
#ScatterPlot between total bill(y) & tips(x).1-Days representation in colors 2-Size of meal represented with size.
sns.scatterplot(
    x="tip",
    y="total_bill",
    data=tips,
    hue="day",
    size="size",
    palette="YlGnBu"
)
plt.title("Tip vs Total Bill (Day-wise & Meal Size)")
plt.xlabel("Tip Amount")
plt.ylabel("Total Bill")
plt.show()
```

```python
#Displaying DataSet using shape,column & head.
iris.shape
iris.columns
iris.head(10)
```

```python
#Histogram
sns.histplot(iris['sepal_length'], kde=True, bins=20, color='darkred')
```

```python
#Box Plot
sns.boxplot(
    x="species",
    y="sepal_width",
    data=iris,
    palette="YlGn"
)
plt.title("Sepal Width Distribution by Species")
plt.show()
```
